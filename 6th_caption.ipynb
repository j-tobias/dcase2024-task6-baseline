```python
from openai import OpenAI
import pandas as pd
from tqdm import tqdm
import json


with open("APIKeys.json", mode="r") as f:
    api_keys = json.load(f)

API_KEY = api_keys["OPENAI"]["SHAREDKEY"]
client = OpenAI(api_key = API_KEY)

def get_caption (caption_1:str , caption_2:str, caption_3:str)-> str:

    prompt = f"""You are given 3 Captions describing the same Audio. Please combine the given captions into a single new improved Caption.
    The goal is to have a new caption, which is semantically similar to the 3 provided, while also being fluent.
    
    Caption 1:
    {caption_1}
    
    Caption 2:
    {caption_2}

    Caption 3:
    {caption_3}

    Give your new caption without any further detailes like 'New Caption:'"""

    response = client.chat.completions.create(
    model="gpt-4", #gpt-4 | gpt-3.5-turbo-16k
    messages=[
        {
        "role": "user",
        "content": prompt
        }
    ],
    temperature=1,
    max_tokens=100,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0
    )

    return response.choices[0].message.content



# LOAD Development csv for the captions
dev_caps = pd.read_csv("data/CLOTHO_v2.1/clotho_csv_files/clotho_captions_development.csv", sep=",")
new_dev_caps = []

for index, row in tqdm(dev_caps.iterrows()):
    caption_1 = row[1]
    caption_2 = row[2]
    caption_3 = row[3]

    new_caption = get_caption(caption_1, caption_2, caption_3)

    new_dev_caps.append(new_caption)

    with open("new_captions_dev.txt", mode="w") as f:
        f.writelines(new_dev_caps)

dev_caps["caption_6"] = new_dev_caps
dev_caps.to_csv("data/CLOTHO_v2.1/clotho_csv_files/clotho_captions_development.csv", sep=",", index=False)
#---------------------------------
eval_caps = pd.read_csv("data/CLOTHO_v2.1/clotho_csv_files/clotho_captions_evaluation.csv", sep=",")
new_eval_caps = []

for index, row in tqdm(eval_caps.iterrows()):
    caption_1 = row[1]
    caption_2 = row[2]
    caption_3 = row[3]

    new_caption = get_caption(caption_1, caption_2, caption_3)

    new_eval_caps.append(new_caption)

    with open("new_captions_eval.txt", mode="w") as f:
        f.writelines(new_eval_caps)

eval_caps["caption_6"] = new_eval_caps
eval_caps.to_csv("data/CLOTHO_v2.1/clotho_csv_files/clotho_captions_evaluation.csv", sep=",", index=False)
#---------------------------------
val_caps = pd.read_csv("data/CLOTHO_v2.1/clotho_csv_files/clotho_captions_validation.csv", sep=",")
new_val_caps = []

for index, row in tqdm(val_caps.iterrows()):
    caption_1 = row[1]
    caption_2 = row[2]
    caption_3 = row[3]

    new_caption = get_caption(caption_1, caption_2, caption_3)

    new_val_caps.append(new_caption)

    with open("new_captions_val.txt", mode="w") as f:
        f.writelines(new_val_caps)

val_caps["caption_6"] = new_val_caps
val_caps.to_csv("data/CLOTHO_v2.1/clotho_csv_files/clotho_captions_validation.csv", sep=",", index=False) 
```
